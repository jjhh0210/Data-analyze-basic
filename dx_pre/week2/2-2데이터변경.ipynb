# 컬럼생성


```python
# 등급 컬럼생성
df['등급'] = 'A' #새로운 컬럼 명시 후 값 부여
display(df)
```

```python
# 성적 컬럼 추가
grade = np.array(['B','D','B','C','A'])
#grade = ['B','D','B','C','A']   #이렇게 해도 작동 o
df['등급'] = grade
display(df)
```

```python
#나이 컬럼 추가
age = pd.Series([21,20,24,23,28])
df['나이'] = age
display(df)
```

# 인덱스 설정 및 리셋

```python
# df 인덱스 확인 (0부터 5 전까지 하나씩 증가한다는 의미)
df.index
```

```python
# df 인덱스 변경해보기 -> df.index로 접근
df.index = ['one','two','three','four','five']
display(df)
```

```python
sex = pd.Series(['여','여','남','남','남'])
df['성별'] = sex
display(df)
```

```python
# 왜 결측치인 NaN이 나올까? dataframe과의 인덱스 타입이 달라서!!
print('시리즈 sex의 인덱스 -->',sex.index)
print('데이터 프레임 df의 인덱스 -->',df.index)
```

```python
# 인덱스 통일 후 다시 성별 컬럼에 값 부여
sex.index = df.index # 성별 시리즈의 인덱스를 df 인덱스로 변경
df['성별'] = sex
df
```

```python
# 현재 설정 되어 있는 인덱스 리셋
#df.reset_index() #여기서 보여지는거는 view이므로 원본에서는 index 컬럼 실제로 없음
df.reset_index(inplace=True) #인수값주면 df 원본에 바로 적용됨
```

```python
df
```

```python
# 인덱스 설정
# df.set_index('이름')
df.set_index('index',inplace=True)
```

# 컬럼 삭제

```python
# 성별 컬럼 삭제(dataFrame의 drop(컬럼명,axis=1))
df.drop('성별',axis=1) # drop에서 axis=1이란 컬럼을 삭제하겠다는 의미!
```
